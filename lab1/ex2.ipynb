```python
import matplotlib.pyplot as plt
import numpy as np
```

```python
# a

t = np.linspace(0, 1/400, 1600) 
plt.stem(t, np.sin(2 * np.pi * 400 * t))
plt.title("Semnal sinusoidal 400Hz")
```

```python
# b

t = np.linspace(0, 3, 3 * 800) 
plt.stem(t, np.sin(2 * np.pi * 800 * t))
plt.title("Semnal sinusoidal 800Hz")
```
