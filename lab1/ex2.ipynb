```python
import matplotlib.pyplot as plt
import numpy as np
```

```python
# a

timp = np.linspace(0, 16, 1600) 
plt.figure(figsize=(20,5))
plt.stem(timp, np.sin(2 * np.pi * 400 * timp))
plt.title("Semnal sinusoidal 400Hz")
```

```python
# b

timp = np.linspace(0, 3) 
plt.figure(figsize=(20,5))
plt.plot(timp, np.sin(2 * np.pi * 800 * timp))
plt.title("Semnal sinusoidal 800Hz")
```

```python
# c
# https://www.tutorialspoint.com/plotting-a-sawtooth-wave-using-matplotlib
timp = np.linspace(0, 5/240) 
plt.figure(figsize=(20,5))
plt.plot(timp, 240 * timp - np.floor(240 * timp))
plt.title("Semnal sawtooth 240Hz")
```

```python
#d
timp = np.linspace(0, 2/300) 
plt.figure(figsize=(20,5))
plt.plot(timp, np.sign(np.sin(2 * np.pi * 300 * timp)))
plt.title("Semnal square 300Hz")
```

```python
#e
I = np.random.rand(128,128)
plt.imshow(I)
plt.title("Semnal 2D aleator")
```

```python
#f
I = np.ones((128,128))

for i in range(128):
    for j in range(128):
        if (i%2==0 and j%2==0):
            I[i, j] = 0

plt.imshow(I)
plt.title("Semnal 2D")
```
