```python
import matplotlib.pyplot as plt
import numpy as np
```

```python
# a

timp = np.linspace(0, 16, 1600) 
plt.figure(figsize=(20,5))
plt.stem(timp, np.sin(2 * np.pi * 400 * timp))
plt.title("Semnal sinusoidal 400Hz")
```

```python
# b

timp = np.linspace(0, 3) 
plt.figure(figsize=(20,5))
plt.plot(timp, np.sin(2 * np.pi * 800 * timp))
plt.title("Semnal sinusoidal 800Hz")
```

```python
# c
# https://www.tutorialspoint.com/plotting-a-sawtooth-wave-using-matplotlib
timp = np.linspace(0, 3) 
plt.figure(figsize=(20,5))
plt.plot(timp, 240 * timp - np.floor(240 * timp))
plt.title("Semnal sawtooth 240Hz")
```
