```python
import matplotlib.pyplot as plt
import numpy as np
```

```python
x = lambda t: np.cos(520 * np.pi * t + np.pi / 3)
y = lambda t: np.cos(280 * np.pi * t + np.pi / (-3))
z = lambda t: np.cos(120 * np.pi * t + np.pi / 3)
```

```python
# a + b

interval = np.linspace(0, 0.1, 500)
fig, axs = plt.subplots(3)
fig.suptitle("Semnale continue")
axs[0].plot(interval, x(interval))
axs[1].plot(interval, y(interval))
axs[2].plot(interval, z(interval))

fig.set_figheight(10)
fig.set_figwidth(20)

axs[0].set_title("x(t)")
axs[1].set_title("y(t)")
axs[2].set_title("z(t)")
```

```python
# c

interval = np.linspace(0, 0.1, 20)
fig, axs = plt.subplots(3)
fig.suptitle("Semnale discrete")
axs[0].stem(interval, x(interval))
axs[1].stem(interval, y(interval))
axs[2].stem(interval, z(interval))

fig.set_figheight(10)
fig.set_figwidth(20)

axs[0].set_title("x[n]")
axs[1].set_title("y[n]")
axs[2].set_title("z[n]")
```
