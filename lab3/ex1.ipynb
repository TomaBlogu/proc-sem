```python
import math
import matplotlib.pyplot as plt
import numpy as np
import time
```

```python
N = 8

fig, axs = plt.subplots(N, figsize=(20,10))
fig.suptitle("DFT de mana")

start_time = time.time()

for n1 in range(N):
    val = []
    for n2 in range(N):
        val.append(math.e**(-2 * np.pi * 1j * n1 * n2 / N))
    axs[n1].plot([col for col in range(N)], [val[col].real for col in range(N)])
    axs[n1].plot([col for col in range(N)], [val[col].imag for col in range(N)], ".--")

print("--- %s seconds ---" % (time.time() - start_time))
```
