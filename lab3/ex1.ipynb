```python
import math
import matplotlib.pyplot as plt
import numpy as np
import time
```

```python
N = 8
X = []

start_time = time.time()

for n1 in range(N):
    a = []
    for n2 in range(N):
        a.append(math.e**(2j * math.pi * n1 * n2 / N))
    X.append(a)

for n1 in range(N):
    for n2 in range(N):
        print(X[n1][n2], end=" ")
    print()

print("--- %s seconds ---" % (time.time() - start_time))
```

```python
fig, axs = plt.subplots(N, figsize=(5,15))
fig.suptitle('Exercitiul 1')

for n1 in range(N):
    real = []
    imag = []
    for n2 in range(N):
        real.append(X[n1][n2].real)
        imag.append(X[n1][n2].imag)
    axs[n1].plot(real, imag)

```
