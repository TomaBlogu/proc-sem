```python
from scipy import misc, ndimage
import numpy as np
import matplotlib.pyplot as plt
```

```python
#1
def semnal(n1, n2):
    return np.sin(2 * np.pi * n1 + 3 * np.pi * n2)

n1, n2 = np.meshgrid(np.arange(0, 10), np.arange(0, 10))

X = semnal(n1, n2)

plt.imshow(X, cmap=plt.cm.gray)
plt.title('$x_{n_1,n_2} = \sin(2\pi n_1 + 3\pi n_2)$')
plt.xlabel('$n_1$')
plt.ylabel('$n_2$')
plt.show()

Y = np.fft.fft2(X)
freq_db = 20*np.log10(abs(Y))

plt.imshow(freq_db)
plt.title('Magnitude Spectrum')
plt.colorbar()
plt.show()
```

```python
#2
def semnal(n1, n2):
    return (np.sin(4 * np.pi * n1) + np.cos(6 * np.pi * n2))

n1, n2 = np.meshgrid(np.arange(0, 10), np.arange(0, 10))

X = semnal(n1, n2)

plt.imshow(X, cmap=plt.cm.gray)
plt.title('$x_{n_1,n_2} = \sin(2\pi n_1 + 3\pi n_2)$')
plt.xlabel('$n_1$')
plt.ylabel('$n_2$')
plt.show()

Y = np.fft.fft2(X)
freq_db = 20*np.log10(abs(Y))

plt.imshow(freq_db)
plt.title('Spectru')
plt.colorbar()
plt.show()
```

```python
#3
Y = np.zeros((10, 10))
Y[0, 5] = 1
Y[0, 10 - 5] = 1
freq_db = 20*np.log10(abs(Y))

plt.imshow(freq_db)
plt.title('$Y_{0,5} = Y_{0,N-5} = 1\text{, altfel }Y_{m_1,m_2} = 0,\ \forall m_1, m_2$')
plt.colorbar()
plt.show()

X = np.fft.ifft2(Y)
X = np.real(X)

plt.imshow(X, cmap=plt.cm.gray)
plt.title('Imagine')
plt.xlabel('$n_1$')
plt.ylabel('$n_2$')
plt.show()
```
