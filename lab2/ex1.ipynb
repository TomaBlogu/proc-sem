```python
import scipy.io.wavfile
import scipy.signal
import matplotlib.pyplot as plt
import numpy as np
```

```python
def sinx(t):
    return 2*np.sin(2*np.pi*400*t)

def cosx(t):
    return 2*np.cos(2*np.pi*400*t - np.pi/2 ) # scadem cu pi/2 pentru a muta semnalul la dreapta
```

```python
xsin = np.arange(0,0.003*np.pi,0.00001)
ysin = sinx(xsin)
```

```python
xcos = np.arange(0,0.003*np.pi,0.00001)
ycos = cosx(xcos)
```

```python
plt.figure(figsize=(20,5))
plt.subplot(1, 2, 1)
plt.plot(xsin,ysin)
plt.subplot(1, 2, 2)
plt.plot(xcos,ycos)
```
