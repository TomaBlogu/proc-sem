```python
import scipy.signal
import matplotlib.pyplot as plt
import numpy as np
```

```python
x1 = lambda t: np.sin(2 * np.pi * 400 * t) # A=1 f=400 faza=0

x2 = lambda t: np.cos(2 * np.pi * 400 * t - np.pi / 2) # scadem cu pi/2 pentru a muta semnalul la dreapta
```

```python
timp = np.linspace(0, 1, 200)
fig, axs = plt.subplots(2)
fig.suptitle("Sin si cos")
axs[0].plot(timp, x1(timp))
axs[1].plot(timp, x2(timp))

fig.set_figheight(10)
fig.set_figwidth(20)

axs[0].set_title("sin(t)")
axs[1].set_title("cos(t)")
```
